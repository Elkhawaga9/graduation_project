```python
import json
from pathlib import Path
import numpy as np
from langchain_openai import OpenAIEmbeddings
```

```python
# -----------------------------
# 1️⃣ List available JSON files
json_dir = Path(r"E:/graduation_project/json_llm_responses")
print(json_dir.exists())  # Should print True

json_files = list(json_dir.glob("*.json"))
print("Available JSON files:")
for i, f in enumerate(json_files, 1):
    print(f"{i}. {f.name}")

```

```python
# -----------------------------
# 2️⃣ Choose a file
choice = int(input("Select JSON file number to generate embeddings: "))
selected_file = json_files[choice - 1]
print(f"Selected: {selected_file.name}")
```

```python
# -----------------------------
# 3️⃣ Load chunks
with open(selected_file, "r", encoding="utf-8") as f:
    chunks = json.load(f)
```

```python
print(chunks[2]['text'])
```

```python
# -----------------------------
# 4️⃣ Extract texts for embedding
texts = [chunk["text"] for chunk in chunks]  # or chunk["llm_response"] if you prefer
ids = [chunk["id"] for chunk in chunks]

```

```python
# -----------------------------
# 5️⃣ Initialize embeddings model
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

text = "This is a test document."
query_result = embeddings.embed_query(text)

# show only the first 100 characters of the stringified vector
print(str(query_result)[:100] + "...")
# Initialize the embeddings model

```

```python
# -----------------------------
# 6️⃣ Generate embeddings
print("Generating embeddings...")
embeddings = embeddings.embed_documents(texts)

```

```python
from pathlib import Path

# Create embeddings directory in the project root
embeddings_dir = Path("../embeddings")  # relative to notebooks/ if running from there
embeddings_dir.mkdir(parents=True, exist_ok=True)
```

```python
# -----------------------------
# 7️⃣ Save embeddings
# Save embeddings for this JSON file
output_file = embeddings_dir / f"{selected_file.stem}_embeddings.npy"
np.save(output_file, np.array(embeddings))
print(f"✅ Saved embeddings to {output_file}")
```
